```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../data/raw/Chicago_Crime.csv")
df = df.sort_values("Date", ascending=False)
df = df.head(500000)
```

```python
df.shape
df.info()
df.describe()
df.isnull().sum()
```

```python
df['Primary Type'].value_counts().plot(kind='bar', figsize=(12,5))
```

```python
plt.scatter(df['Longitude'], df['Latitude'], alpha=0.1)
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df['Hour'] = df['Date'].dt.hour
df['Day'] = df['Date'].dt.day_name()
df['Month'] = df['Date'].dt.month
```

```python
sns.countplot(x='Arrest', data=df)
sns.countplot(x='Domestic', data=df)
```
